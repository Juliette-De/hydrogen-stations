```python
import pandas as pd 
```

```python
df = pd.read_csv('tmja2018.csv', sep="\t")
df
```

```python
df = df.dropna()
df['ratioPL'] = df['ratioPL'].str.replace(",", ".").astype(float)
df = df[df['ratioPL'] >= 0]
df 
```

```python
df['xD'] = df['xD'].str.replace(",", ".").astype(float)
df['yD'] = df['yD'].str.replace(",", ".").astype(float)
df['xF'] = df['xF'].str.replace(",", ".").astype(float)
df['yF'] = df['yF'].str.replace(",", ".").astype(float)
```

```python
#!pip install pyproj
import pandas as pd
import pyproj

# Define the Lambert 93 and WGS 84 coordinate reference systems
lambert93 = pyproj.Proj(init='EPSG:2154')
wgs84 = pyproj.Proj(init='EPSG:4326')


# Define a lambda function to transform each row of coordinates
transform_coords_D = lambda row: pyproj.transform(lambert93, wgs84, row['xD'], row['yD'])
transform_coords_F = lambda row: pyproj.transform(lambert93, wgs84, row['xF'], row['yF'])

# Apply the lambda function to each row of the DataFrame
df[['lonD', 'latD']] = df.apply(transform_coords_D, axis=1, result_type='expand')
df[['lonF', 'latF']] = df.apply(transform_coords_F, axis=1, result_type='expand')

# Print the resulting DataFrame
print(df)
```

```python
df['truck_traffic'] = df['TMJA'] * df['ratioPL'] / 100
#df.groupby(['binLonD', 'binLatD'])['truck_traffic'].mean().dropna().reset_index()
```

```python
region_dep = pd.read_csv("departements-region.csv", sep=';')
```

```python
df["depPrD"] = df["depPrD"].astype(str)
```

```python
df = df.merge(region_dep[["num_dep", "region_name"]], right_on="num_dep", left_on="depPrD", how="left")
```

```python
df
```

```python
df_by_region_and_road = df.groupby(["route", "region_name"])["truck_traffic"].mean().reset_index()
```

```python
df = df.merge(df_by_region_and_road, how='left', on=["region_name", "route"])
df.rename(columns = {"truck_traffic_y" : "truck_traffic_region", "truck_traffic_x" : "truck_traffic_segment"}, inplace=True)
```

Idée : faire un K Means pour scinder les coordonnées des auto routes en 55 clusters, et on aggrège les données de traffic sur ces klusters. 

```python
from sklearn.cluster import KMeans
```

```python
kMeans = KMeans(55)
kMeans.fit(df[['lonD', 'latD']])
```

```python
df['cluster'] = kMeans.labels_
df
```

```python
import seaborn as sns
import numpy as np

# set the number of colors you want to generate
n = 10000

# generate a random list of n hexadecimal colors
colors = ['#'+''.join([np.random.choice(list('0123456789ABCDEF')) for j in range(6)]) for i in range(n)]

```

```python
#df_traffic = df.groupby(['region_name'])['truck_traffic'].mean().reset_index()
#df_traffic.rename(columns = {'truck_traffic':'avg_traffic'},inplace=True)
#df = df.merge(df_traffic, on='cluster')
df['avg_traffic_seg_bin'] = pd.cut(df['truck_traffic_segment'], 20, labels=np.linspace(1, 20, 20))
df
```

```python
import folium
import matplotlib.pyplot as plt

# create a folium map centered on the US
#map_center = [39.8333, -98.5855]
m = folium.Map(zoom_start=4)

# add markers for starting and ending positions

j=0
for route in list(df.route.unique()):
  for i, row in df[df['route'] == route].iterrows():
    folium.PolyLine(
        locations=[[row['latD'], row['lonD']], [row['latF'], row['lonF']]],
        color=colors[j],
        tooltip=row['route'],
        weight=row['avg_traffic_seg_bin']
    ).add_to(m)
  j += 1

m  # display the map
```
